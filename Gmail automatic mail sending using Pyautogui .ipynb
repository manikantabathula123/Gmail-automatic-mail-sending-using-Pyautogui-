```python
import pyautogui as py # Importing the PYAUTOGUI module

import time #Importing time module to delay the excution time between the commands.

time.sleep(4)

# Here, png's are snippets images of desired task to activate. So, if we try to run we need to ensure that the respective images
# are there in the default location of the python file.

x=py.locateCenterOnScreen(r"google.PNG",confidence=0.9) # to take the co-ordinates of the google chrome browser 

py.click(x) # to click on the google chrome browser.

py.moveTo(277,646,3) #to move mouse to the (277,646) coordinates within a time of 3 seconds. 

py.leftClick()

# Once webpage loaded we need to enter our destination website.

py.write("https://www.gmail.com/",interval=0.25) # to type webaddress with a interval of 0.25 between each letter.

time.sleep(1) 

# New few steps are just login process.

py.press("enter") 

time.sleep(2)

x=py.locateCenterOnScreen(r"login_email.png",confidence=0.5) 

py.click(x)

time.sleep(3)

py.write('enter mail address')

time.sleep(2)

x=py.locateCenterOnScreen(r"next.png",confidence=0.5) 

py.click(x)

time.sleep(3)

py.write('enter_password_here')

py.click(x)

time.sleep(2)

x=py.locateCenterOnScreen(r"next.png",confidence=0.5) 

py.click(x)

# By now we can able to view our mail. Next few steps are similiar to what we just done for mail login.

time.sleep(15)

x=py.locateCenterOnScreen(r"compose.png",confidence=0.5) 

py.click(x)

time.sleep(2)

py.write("enter recepients mail adress ")

time.sleep(2)

py.press(["tab","tab"])

py.write("gratitude")

py.press(["tab"])

time.sleep(2)

py.write("write_subject_here")

time.sleep(2)

x=py.locateCenterOnScreen(r"send.png",confidence=0.9) 

py.click(x)

#Successfully, by end of this step we can finally send mail.
```
